```python
import psycopg2
import pandas as pd

# Database connection configuration
DB_CONFIG = {
    'dbname': 'telegram_data',
    'user': 'postgres',
    'password': 'youwork',
    'host': 'localhost',
    'port': 5432
}

# CSV file path for exporting data
EXPORT_CSV = "exported_data.csv"

# Connect to PostgreSQL
def connect_to_db():
    try:
        conn = psycopg2.connect(**DB_CONFIG)
        print("Database connection successful.")
        return conn
    except Exception as e:
        print(f"Error connecting to database: {e}")
        return None

# Export data from the database to CSV and return as DataFrame
def export_data_to_csv(conn, query, csv_file):
    try:
        # Use pandas to fetch data and export it to CSV
        df = pd.read_sql_query(query, conn)
        
        # Save to CSV file
        df.to_csv(csv_file, index=False)
        print(f"Data successfully exported to {csv_file}")
        
        # Return the DataFrame
        return df
    except Exception as e:
        print(f"Error exporting data to CSV: {e}")
        return None

# Main function to load and export data
if __name__ == "__main__":
    conn = connect_to_db()
    if conn:
        # Define the SQL query to fetch data
        query = "SELECT * FROM telegram_data"
        
        # Export data to CSV and get the DataFrame
        df = export_data_to_csv(conn, query, EXPORT_CSV)
        
        if df is not None:
            print("Data loaded into DataFrame:")
            print(df.head())  # Display first 5 rows for verification
        
        # Close the database connection
        conn.close()
```
